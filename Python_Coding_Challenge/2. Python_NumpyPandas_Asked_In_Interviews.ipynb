```python
#### Q15. Write a function that takes a region name as input and returns:   
'''
- Total sales
- Average sales
- Number of unique customers
Hint:   
Use filtering + aggregation inside a function
'''
```

```python
def get_region_stats(region):
  df_region = df[df['region'] == region]
  return df_region['sales'].sum(), df_region['sales'].mean(), df_region['customer_name'].nunique()
```
